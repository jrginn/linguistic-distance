```python
'''
Input: directory name of txt files
Output: string containing all text of all files
'''
def aggregate_text(filepath):
    str = ''
    for filename in os.listdir(filepath):
        with open((filepath + '/'+ filename), 'r') as file:
            print(filepath + '/'+ filename)
            str += file.read().replace('\n', ' ')
        str += '\n'
    return str.lower()
```

```python
def str_to_freq(str):
    tokens = [wordpunct_tokenize(str)]
    words = [w for w in tokens[0] if w not in punctuation]
    frequency = defaultdict(int)
    for w in words:
        frequency[w] += 1
    return frequency
```
